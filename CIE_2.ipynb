```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Titanic-Dataset.csv')
meain_age = df['Age'].mean()
mode_age = df['Age'].mode()[0]
median_age = df['Age'].median()
meain_fare = df['Fare'].mean()
mode_fare = df['Fare'].mode()[0]
median_fare = df['Fare'].median()
print('Mean of Age :', meain_age)
print('Mode of Age:', mode_age)
print('Median of Age:', median_age)
print('Mean of Fare:', meain_fare)
print('Mode of Fare:', mode_fare)
print('Median of Fare:', median_fare)
```
